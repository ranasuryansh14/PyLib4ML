```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Importing Dataset
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
##Plot Total sales By Product
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y='Units Sold',data=sales_df,estimator=sum)
plt.title("Total Sales by Product")
plt.show()
```

```python
total_units_sold = sales_df['Units Sold'].sum()
```

```python
total_units_sold
```

```python
##Plot Total revenue Generated Ffrom Different Platform
plt.figure(figsize=(10,6))
sns.barplot(x='Payment Method',y='Total Revenue',data=sales_df,estimator=sum)
plt.title("Total Sales by Product")
plt.show()
```

```python
total_revenue_by_region = sales_df.groupby('Region')['Total Revenue'].sum()
print("Total Revenue by Region:")
print(total_revenue_by_region)
```

```python
# Given data
data = {
    'Region': ['Asia', 'Europe', 'North America'],
    'Total Revenue': [22455.45, 21268.06, 36844.34]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot bar graph
plt.figure(figsize=(10, 6))
plt.bar(df['Region'], df['Total Revenue'], color=['blue', 'lightgreen', 'teal'])
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.title('Total Revenue by Region')
plt.show()
```
