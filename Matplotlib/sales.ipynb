### Sales Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
# data Frame
sale_data_df = pd.read_csv('sales_data.csv')
```

```python
sale_data_df.head()
```

### Grouping 

```python
sale_data_df.info()
```

```python
## Plot Total sales by products
total_sales_by_product = sale_data_df.groupby('Product Category')['Total Revenue'].sum()
print(total_sales_by_product)
```

##### Ploting Bar plot

```python
total_sales_by_product.plot(kind = "bar", color = "teal")
```

```python
## Plot sales trend over time
sales_trend = sale_data_df.groupby('Date')['Total Revenue'].sum().reset_index() ## why reset index we will get 2 coulumns used to plot in x y axis
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```

```python
## region wise trend
total_sales_by_region = sale_data_df.groupby('Region')['Total Revenue'].sum()
print(total_sales_by_region)
```

```python
# ploting in pie 
explode = [0,0.05,0]
plt.figure(figsize=(6, 6))
plt.pie(total_sales_by_region,explode=explode, labels=total_sales_by_region.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Sales by Region')
plt.show()
```
