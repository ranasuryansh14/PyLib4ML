### __Data Visualisation with Matplotlib__

### Matplotlib

It is powerful plotting library for python that enables the creation of static, animated and interactive visualization. It is widely used for data visualization in data science and analytics.

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Creating a line plot
plt.plot(x,y) # Ploting The Co-ordinates
plt.xlabel('X axis')# Label in x axis
plt.ylabel('Y axis')# Lable in y axis
plt.title("Basic Line Plot") # Tittle Label
plt.show()# show Command
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

##  Creating Customized Plot

plt.plot(x,y,color='green', linestyle="--", marker ='o', linewidth =3, markersize=9 ) # marker for cordinate
# supported line style are --> -,--,-.,:
plt.grid(True)# graph
```

```python
## Creating Multi plot
#Sample Data
x = [1,2,3,4,5]
y = [1,4,9,16,25]
z = [1,2,3,4,5]
w = [4,5,6,8,10]
plt.figure(figsize=(9,5))
plt.subplot(2,2,1) # Rows and Column ( 1 row and 2 col)
plt.plot(x,y , color="green")
plt.title("Plot 1")

plt.subplot(2,2,2) # Rows and Column ( 1 row and 2 col)
plt.plot(y,z , color="red")
plt.title("Plot 2")

plt.subplot(2,2,3) # Rows and Column ( 1 row and 2 col)
plt.plot(x,z , color="blue")
plt.title("Plot 3")

plt.subplot(2,2,4) # Rows and Column ( 1 row and 2 col)
plt.plot(w,z , color="yellow")
plt.title("Plot 4")
```

### Bar Plot 

#### categories

```python
categories = [2024,2025,2026,2027,2028,2029]
values = [1,1,2,3,4,5]

# Creating  bar
plt.bar(categories,values,color = 'red')
plt.xlabel("Years")
plt.ylabel("Houses")
plt.title("Data")
plt.show()
```

### Histogram

```python
#data 
data = [1,2,3,4,4,5,6,6,7,6,4,4,9,5,4,3,2]

## create a histogram
plt.hist(data,bins = 5,color ="orange",edgecolor="black")
```

### Scatter Plot


```python
# data 
x = [1,2,3,4,5,6]
y=[2,4,6,9,0,8]

plt.scatter(x,y,color = "blue", marker='*')
```

### PyChart

```python
#labels
labels = ["A","B","C","D"]
sizes = [30,20,10,40]
colors = ['gold','yellowgreen','lightgreen','lightblue']
explode = [0,0,0.4,0] ## cut out the part 

plt.pie(sizes,explode=explode,labels=labels, colors=colors, autopct = "%1.1f%%",shadow=True)
```
